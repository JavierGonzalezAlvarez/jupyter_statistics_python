# empty values - fill with the mean - SimpleImputer

```python
from sklearn.impute import SimpleImputer 
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer=imputer.fit(X[:,5:])
X[:,5:]=imputer.transform(X[:,5:])
imputation = X[:,5:]
imputation
```

# reshape, numpy

```python
np_array = np.array(imputation).reshape(1,540)
np_array
```

# new array with for

```python
list = []          
for i in range(len(np_array)) :  
    for j in range(len(np_array[i])) :  
        list.append(np_array[i][j])        
list
```

# count numbers by intervals

```python
data = np.array(np_array)
interval = [0.0, 10.00, 20.00, 30.00, 50.00, 80.00, 100.00]

count, interval_edges = np.histogram(data, interval)

print(f"INTERVAL     COUNT")
print("-------------------")

for low, hight, count in zip(interval_edges, np.roll(interval_edges, -1), count):
    print(f"{f'{low}-{hight}': <10}    {count}")
```
